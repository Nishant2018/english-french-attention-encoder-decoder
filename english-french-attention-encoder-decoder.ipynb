     What Is Seq2Seq Model ?
</div>

The Sequence-to-Sequence (Seq2Seq) model is a type of neural network architecture used for sequence-to-sequence learning tasks, particularly in natural language processing (NLP) and machine translation. It's designed to handle input and output sequences of variable lengths, making it suitable for tasks like language translation, text summarization, and question answering.

Here's an explanation of how Seq2Seq models work:

1. **Encoder-Decoder Architecture**:
   - The Seq2Seq model consists of two main components: an encoder and a decoder.
   - The encoder takes the input sequence (e.g., a sentence in one language) and encodes it into a fixed-length vector called the context vector. This vector represents the input sequence's semantic meaning and is used as the initial state for the decoder.

2. **Encoder**:
   - The encoder is typically a recurrent neural network (RNN) or its variations like Long Short-Term Memory (LSTM) or Gated Recurrent Unit (GRU).
   - Each element of the input sequence is fed into the encoder one by one, and the hidden state of the encoder is updated at each time step. The final hidden state or the output of the encoder is used as the context vector.

3. **Decoder**:
   - The decoder is also an RNN or its variants and is conditioned on the context vector produced by the encoder.
   - It takes the context vector as input and generates the output sequence (e.g., a translated sentence in another language) one element at a time.
   - The decoder uses the context vector and the previously generated elements of the output sequence to predict the next element in the sequence.
   - During training, the decoder is fed with the target sequence (the correct translation) to learn to generate the output correctly.

4. **Training**:
   - During training, the model is optimized to minimize a loss function, such as cross-entropy loss, which measures the difference between the predicted sequence and the target sequence.
   - Techniques like teacher forcing, where the model is fed with the actual target sequence during training instead of its own predictions, can be used to stabilize and accelerate training.

5. **Inference**:
   - During inference or testing, the trained model is used to generate sequences without teacher forcing.
   - The decoder starts with a special start-of-sequence token and generates subsequent elements until it produces an end-of-sequence token or reaches a predefined maximum length.

Seq2Seq models have been instrumental in improving the performance of various NLP tasks, especially in machine translation where they have shown significant advancements compared to traditional rule-based or statistical approaches. They are also used in tasks like text summarization, dialogue generation, and more recently, in multimodal tasks involving both text and images or videos.

     Steps (Preprocessing -> Model Creation)
</div>

Here are the steps from preprocessing to model creation for an English to French translation task using a Seq2Seq model:

1. **Data Preprocessing**:
   - Tokenization:
     - Tokenize English and French sentences into words or subword units using libraries like NLTK or Spacy.
   - Padding and Truncation:
     - Pad or truncate sequences to a fixed length to ensure uniform input size for the model.
   - Vocabulary Creation:
     - Build vocabulary dictionaries for both English and French languages, mapping words to unique integer indices.
   - Data Splitting:
     - Split the dataset into training, validation, and test sets.

2. **Word Embeddings**:
   - Initialize word embeddings for English and French words using pre-trained embeddings like Word2Vec, GloVe, or train embeddings from scratch.
   - Map each word in the vocabulary to its corresponding embedding vector.

3. **Encoder-Decoder Model**:
   - Build an Encoder-Decoder architecture for sequence-to-sequence (Seq2Seq) translation.
   - Define an encoder model to process input English sentences and generate encoder hidden states.
   - Define a decoder model to decode the encoder hidden states and generate French translations.

4. **Attention Mechanism**:
   - Implement an attention mechanism to help the decoder focus on relevant parts of the input sequence during decoding.
   - Calculate attention weights based on the similarity between encoder and decoder hidden states.

5. **Training**:
   - Compile the Seq2Seq model with appropriate loss function (e.g., categorical cross-entropy) and optimizer (e.g., Adam).
   - Train the model using the training dataset, passing English sentences as input and corresponding French sentences as target output.
   - Monitor training progress using validation data and adjust hyperparameters if necessary (e.g., learning rate, batch size).

6. **Evaluation**:
   - Evaluate the trained model using the test dataset to measure translation accuracy and performance metrics such as BLEU score.
   - Generate translations for English sentences from the test set and compare them with ground truth French translations.

7. **Inference**:
   - Implement an inference pipeline to translate new English sentences into French using the trained model.
   - Preprocess new input sentences by tokenizing, padding/truncating, and converting words to indices using the pre-built vocabulary.
   - Feed the preprocessed input into the encoder, generate encoder hidden states, and decode using the decoder to obtain the translated output.

8. **Model Deployment**:
   - Deploy the trained Seq2Seq model for English to French translation in production environments using frameworks like TensorFlow Serving or Flask API.
   - Create a user-friendly interface or API endpoint to accept English text inputs and return corresponding French translations.

By following these steps, you can build and deploy an effective Seq2Seq model for English to French translation.

* **Null Value Checking and Removing**

```python
data.isnull().sum()
```

* **LowerCasing**

```python
data['English words/sentences'] = data['English words/sentences'].str.lower()
data['French words/sentences'] = data['French words/sentences'].str.lower()
```

* **Removing HTML**

```python
def removing_html(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'',text)

data['English words/sentences'] = data['English words/sentences'].apply(removing_html)
data['French words/sentences'] = data['French words/sentences'].apply(removing_html)
```

* **Removing URL's**

```python
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'',text)
data['English words/sentences'] = data['English words/sentences'].apply(remove_url)
data['French words/sentences'] = data['French words/sentences'].apply(remove_url)
```

* **Removing the Alphanumeric Characters**

```python
def preprocess_text(text):
    pattern = re.compile(r'[^a-zA-Z0-9\s]')
    return pattern.sub(r'',text)

data['English words/sentences'] = data['English words/sentences'].apply(preprocess_text)
data['French words/sentences'] = data['French words/sentences'].apply(preprocess_text)
```

* **Removing Punctuations**

```python
string.punctuation
```

```python
def remove_punctuations(text):
    exclude_english = set(string.punctuation)
    return ''.join(char for char in text if char not in exclude_english)
data['English words/sentences'] = data['English words/sentences'].apply(remove_punctuations)
data['French words/sentences'] = data['French words/sentences'].apply(remove_punctuations)
```

* **Removing contractions**

```python
def expand_contractions(text):
    expanded_text = contractions.fix(text)
    return expanded_text

data['English words/sentences'] = data['English words/sentences'].apply(expand_contractions)
data['French words/sentences'] = data['French words/sentences'].apply(expand_contractions)
```

* **Tokenize the text**

```python
def do_tokenization(text):
    token_words = word_tokenize(text)
    return token_words

data['English words/sentences'] = data['English words/sentences'].apply(do_tokenization)
data['French words/sentences'] = data['French words/sentences'].apply(do_tokenization)
```

```python
data.sample(5)
```

* **Removing Stop Words**

```python
nltk.download('stopwords')
```

```python
stop_words_english = set(stopwords.words('english'))
stop_words_french = set(stopwords.words('french'))
```

```python
def remove_stopwords(text,language = 'english'):
    if language == 'english':
        filtered_words_english = [word for word in text if word.lower() not in stop_words_english]
        return ' '.join(filtered_words_english)
    elif language == 'french':
        filtered_words_french = [word for word in text if word.lower() not in stop_words_french]
        return ' '.join(filtered_words_french)
    else:
        return ValueError("Unsupported Language")

data['English words/sentences'] = data['English words/sentences'].apply(lambda x: remove_stopwords(x,language = 'english'))
data['French words/sentences'] = data['French words/sentences'].apply(lambda x: remove_stopwords(x,language = 'french'))
```

```python
data.sample(5)
```

* **Stemming the text**

```python
def do_stemming(text):
    ps = PorterStemmer()
    words = text.split()
    return [ps.stem(word) for word in words]
data['English words/sentences'] = data['English words/sentences'].apply(do_stemming)
data['French words/sentences'] = data['French words/sentences'].apply(do_stemming)
```

* **Encoder-Decoder Model**

```python
max_length_input = x_train.shape[1]
max_length_output = y_train.shape[1]
input_vocab_size = len(tok.word_index) + 1
output_vocab_size = len(tok_french.word_index) + 1

y_train_padded = np.concatenate([y_train, np.zeros((len(y_train), 1))], axis=1)
y_test_padded = np.concatenate([y_test, np.zeros((len(y_test), 1))], axis=1)

# Define Encoder model
encoder_inputs = Input(shape=(max_length_input,))
encoder_embedding = Embedding(input_dim=input_vocab_size, output_dim=100)(encoder_inputs)
encoder_lstm = LSTM(256, return_state=True)
encoder_outputs, state_h, state_c = encoder_lstm(encoder_embedding)
encoder_states = [state_h, state_c]

# Define Decoder model
decoder_inputs = Input(shape=(max_length_output,))
decoder_embedding = Embedding(input_dim=output_vocab_size, output_dim=100)(decoder_inputs)
decoder_lstm = LSTM(256, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_embedding, initial_state=encoder_states)
decoder_dense = Dense(output_vocab_size, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)

# Define the full Encoder-Decoder model
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Print model summary
print(model.summary())
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('model_checkpoint.keras', save_best_only=True)  # Updated filepath

# Train the model with callbacks
history = model.fit(
    x=[x_train, y_train_padded[:, :-1]],
    y=y_train_padded[:, 1:],
    batch_size=32,
    epochs=50,
    validation_data=([x_test, y_test_padded[:, :-1]], y_test_padded[:, 1:]),
    callbacks=[early_stopping, model_checkpoint],  
)
```

```python
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, Concatenate, Dot
from tensorflow.keras.models import Model

# Define Encoder model
encoder_inputs = Input(shape=(max_length_input,))
encoder_embedding = Embedding(input_dim=input_vocab_size, output_dim=100)(encoder_inputs)
encoder_lstm = LSTM(256, return_sequences=True, return_state=True)
encoder_outputs, state_h, state_c = encoder_lstm(encoder_embedding)
encoder_states = [state_h, state_c]

# Define Decoder model
decoder_inputs = Input(shape=(max_length_output,))
decoder_embedding = Embedding(input_dim=output_vocab_size, output_dim=100)(decoder_inputs)
decoder_lstm = LSTM(256, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_embedding, initial_state=encoder_states)

# Attention mechanism using Dot product
attention_scores = Dot(axes=[2, 2])([decoder_outputs, encoder_outputs])
attention_weights = Dense(max_length_input, activation='softmax')(attention_scores)
attention_context = Dot(axes=[2, 1])([attention_weights, encoder_outputs])

# Concatenate attention context with decoder outputs
decoder_combined_context = Concatenate()([decoder_outputs, attention_context])

decoder_dense = Dense(output_vocab_size, activation='softmax')
decoder_outputs = decoder_dense(decoder_combined_context)

# Define the full Encoder-Decoder model with attention
model_attention = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile the model
model_attention.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Print model summary
print(model_attention.summary())
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('model_checkpoint_attention.keras', save_best_only=True)  

# Train the model with callbacks
history_attention = model_attention.fit(
    x=[x_train, y_train_padded[:, :-1]],
    y=y_train_padded[:, 1:],
    batch_size=64,
    epochs=15,
    validation_data=([x_test, y_test_padded[:, :-1]], y_test_padded[:, 1:]),
    callbacks=[early_stopping, model_checkpoint],  
)
```
